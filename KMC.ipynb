**Q1.** This question is a case study for $k$ means clustering.

1. Load the `airbnb_hw.csv` data. Clean `Price` along with `Beds`, `Number of Reviews`, and `Review Scores Rating`.
2. Maxmin normalize the data and remove any `nan`'s (`KMeans` from `sklearn` doesn't accept `nan` input).
3. Use `sklearn`'s `KMeans` module to cluster the data by `Beds`, `Number of Reviews`, and `Review Scores Rating` for `k=6`.
4. Use `seaborn`'s `.pairplot()` to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.
5. Use `.groupby` and `.describe` to compute the average price for each cluster. Which clusters have the highest rental prices?
6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
import pandas as pd
import numpy as np
````

````python
#1
df = pd.read_csv('/content/kmc/data/airbnb_hw.csv')
df['price'] = df['Price']
df['beds'] = df['Beds']
df['review'] = df['Number Of Reviews']
df['score'] = df['Review Scores Rating']

temp = df.loc[:,['price','beds','review','score']]

temp['price'] = temp['price'].str.replace(",","")
temp['price'] = pd.to_numeric(temp['price'])
temp['price'].unique()
temp['beds'] = temp['beds'].fillna(1)
temp = temp.dropna()
````

````python
#2
def maxmin(x):
    u = (x-min(x))/(max(x)-min(x))
    return u

temp2 = temp.drop('price',axis=1)
temp2 = temp2.apply(maxmin)
````

````python
#3
from sklearn.cluster import KMeans # Import kmc
import matplotlib.pyplot as plt


model = KMeans(n_clusters=6, max_iter=300, n_init = 10, random_state=0) # Create a model for
model = model.fit(temp2) # Fit the emodel
temp2['cluster'] = model.labels_
temp2.describe()
````

````python
#4
import seaborn as sns
sns.pairplot(data=temp2,hue='cluster')
````

````python
#5
temp['cluster'] = model.labels_
temp.loc[:,['price','cluster'] ].groupby('cluster').describe()

# The clusters with the highest average rental price is cluster one, with a price of almost $300 a night, while the cluster with the highest maximum rental price is cluster 0, though the prices
# there are lower on average than cluster 1
````

````python
#6
k_bar = 15
k_grid = np.arange(1,k_bar+1) # The range of k's to consider
SSE = np.zeros(k_bar) # We'll store the training error here
for k in range(k_bar): # For indices from 0 to k_bar-1
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0) # Create a model for k
    model = model.fit(temp2) # Fit the emodel
    SSE[k] = model.inertia_ # Record the sum of squared error for k in the SSE vector

sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot') # Create scree plot

model = KMeans(n_clusters=3, max_iter=300, n_init = 10, random_state=0) # Create a model for
model = model.fit(temp2) # Fit the emodel
temp2['cluster'] = model.labels_

sns.pairplot(data=temp2,hue='cluster')

temp['cluster'] = model.labels_
temp.loc[:,['price','cluster'] ].groupby('cluster').describe()
````

````python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans # Import kmc
import matplotlib.pyplot as plt
````

````python
#1
def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
#3
def maxmin(x): # Max-min scaler function
    x = (x-min(x))/(max(x)-min(x))
    return x

def scree(data): # I built a function to make scree plots quickly
    # Preprocess the data:
    X = data.loc[ : , ['x1','x2'] ] # Select data to cluster
    X = X.apply(maxmin) # Normalize the data
    # Create a scree plot:
    k_bar = 15
    k_grid = np.arange(1,k_bar+1) # The range of k's to consider
    SSE = np.zeros(k_bar) # We'll store the training error here
    for k in range(k_bar): # For indices from 0 to k_bar-1
        model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0) # Create a model for k
        model = model.fit(X) # Fit the emodel
        SSE[k] = model.inertia_ # Record the sum of squared error for k in the SSE vector
    scree_plot, axes = plt.subplots()
    sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot') # Create scree plot
    axes.set_ylim(0, 35)
scree( data = df0_5 )
#As the noise increases, the curve at the bottom rounds out more, goes from a sharp 90 degrees to a more gradual curve.
````

````python
#4
#As the noise increases, the grouping of the data becomes less distinct. Because of the nature of the scree plot, as the clusters become less distinct, the elbow of the scree plot also
#becomes less distinct. Thus the scree plot makes an eblow with distinct groupings, and doesn't with less distinct groupings.
````
